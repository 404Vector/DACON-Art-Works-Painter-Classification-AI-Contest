```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import glob
import pandas as pd
```

```python
data_lst = glob.glob('../data/test/*.*')
len(data_lst)
```

```python
sizes_h = []
sizes_w = []
for path in data_lst:
    img = cv2.imread(path)
    sizes_h.append(img.shape[0])
    sizes_w.append(img.shape[1])
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 10))
fig.suptitle('Image Size')

ax.set_xlabel('w_size')
ax.set_ylabel('h_size')

ax.scatter(sizes_w, sizes_h, marker='o')
plt.show()
```

```python
print(min(sizes_w), min(sizes_h))
```
