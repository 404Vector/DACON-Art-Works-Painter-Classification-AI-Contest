```python
df = pd.read_csv('../data/train.csv')
df.head()
```

```python
fig, ax = plt.subplots(figsize=(12, 7))
sns.countplot(data=df, x='artist', ax=ax)
plt.xticks(rotation=90)
plt.show()
```

```python
g = df.groupby('artist', group_keys=False)
```

```python
sample = pd.DataFrame()
for _ in range(10):
    sample = pd.concat([sample, g.apply(lambda x: x.sample(21))])
```

```python
sample.sample(10)
```

```python
fig, ax = plt.subplots(figsize=(12, 7))
sns.countplot(data=sample, x='artist', ax=ax)
plt.xticks(rotation=90)
plt.show()
```
